```python
import bls 
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import numpy as np

from blsconnect import RequestBLS
```

```python
from config import BLS_API_KEY
bls = RequestBLS(key=BLS_API_KEY)

Boone_Co_unemployment_monthly_stats = pd.DataFrame(bls.series(['LAUCN180110000000003','LAUCN180110000000004','LAUCN180110000000005','LAUCN180110000000006'], 2008, 2008))
```

```python
Boone_Co_unemployment_monthly_stats
```

```python
#Renamed the column names
Boone_Co_unemployment_monthly_stats.columns = ['year','period','periodName','Unemployment Rate', 'Unemployed', 'Employed','Labor Force']
```

```python
#Boone_Co_unemployment_monthly_stats.index.names = ['Date']
#Boone_Co_unemployment_monthly_stats.index.names
Boone_Co_unemployment_monthly_stats
```

```python
new_df_Boone = Boone_Co_unemployment_monthly_stats
# Create Year-Month by adding columns year and periodName
new_df_Boone['Year-Month'] = new_df_Boone["year"].astype(str) + new_df_Boone["periodName"] 
new_df_Boone['County'] = 'Boone'

new_df_Boone.head()
new_df_Boone.to_html("BooneData.html")
```

```python


x_axis = new_df_Boone["periodName"]
y_axis = new_df_Boone["Unemployed"]
tick_locations = [value for value in x_axis]                   
qp11 = plt.bar(x_axis, y_axis, align="center", width =.7)

plt.xticks(tick_locations, new_df_Boone["periodName"] , rotation = 'vertical')
plt.xlabel('Months Jan - Feb',  fontsize=20)
plt.ylabel('Unemployed',  fontsize=20)
plt.title("2008 Unemployment Chart", fontsize=20)
plt.legend( qp11,x_axis,loc='upper center', fancybox=True, shadow=True, bbox_to_anchor=(1.30, 1.0))
plt.savefig('Umemployment2008.png', bbox_inches = 'tight')
plt.show()
```
