```python
boone_lake_warrick_covid = boone_lake_warrick_covid.melt(id_vars=['County', 'GEOGID'], var_name='Date', value_name="Count")
boone_lake_warrick_covid['Date'] = boone_lake_warrick_covid['Date'].map(lambda x: x.lstrip('dD_'))
```

```python
boone_lake_warrick_covid
```

```python

import numpy as np
boone_covid = boone_lake_warrick_covid.loc[boone_lake_warrick_covid['County'].isin(['Boone County'])]
boone_covid = boone_covid.reset_index(drop=True) 
boone_covid.replace('', np.nan, inplace=True)
boone_covid.fillna(0, inplace=True)
boone_covid['Count'] = boone_covid['Count'].astype(int)
```

```python
boone_covid
```

```python
lake_covid = boone_lake_warrick_covid.loc[boone_lake_warrick_covid['County'].isin(['Lake County'])]
lake_covid = lake_covid.reset_index(drop=True) 
lake_covid.replace('', np.nan, inplace=True)
lake_covid.fillna(0, inplace=True)
lake_covid['Count'] = lake_covid['Count'].astype(int)
lake_covid
```

```python
warrick_covid = boone_lake_warrick_covid.loc[boone_lake_warrick_covid['County'].isin(['Warrick County'])]
warrick_covid = warrick_covid.reset_index(drop=True) 
warrick_covid.replace('', np.nan, inplace=True)
warrick_covid.fillna(0, inplace=True)
warrick_covid['Count'] = warrick_covid['Count'].astype(int)
warrick_covid
```

```python
import pickle
boone_covid.to_pickle("boone_covid.pkl")
lake_covid.to_pickle("lake_covid.pkl")
warrick_covid.to_pickle("warrick_covid.pkl")
```

```python
def allfridays(year):
    return pd.date_range(start=str(year), end=str(year+1), 
                         freq='W-FRI').strftime('%Y%m%d').tolist()
```

```python
Boone_weekly_claims = pd.read_csv("sample_data/BooneCounty_Claims.csv")
Boone_weekly_claims['Date'] = allfridays(2020)[:14]
Boone_weekly_claims['Claims'] = Boone_weekly_claims['Claims'].astype(int)

Boone_weekly_claims
```

```python
Lake_weekly_claims = pd.read_csv("sample_data/LakeCounty_Claims.csv")
Lake_weekly_claims['Date'] = allfridays(2020)[:14]
Lake_weekly_claims['Claims'] = Lake_weekly_claims['Claims'].map(lambda x: x.replace(',', ''))

Lake_weekly_claims['Claims'] = Lake_weekly_claims['Claims'].astype(int)

Lake_weekly_claims
```

```python
Warrick_weekly_claims = pd.read_csv("sample_data/WarrickCounty_Claims.csv")
Warrick_weekly_claims['Date'] = allfridays(2020)[:14]
Warrick_weekly_claims['Claims'] = Warrick_weekly_claims['Claims'].astype(int)

Warrick_weekly_claims
```

```python
boone_covid_claims = pd.merge(Boone_weekly_claims[['Claims', 'Date']], 
                                           boone_covid[['Count', 'Date']], how='outer', on='Date', sort=True)
boone_covid_claims.fillna(0, inplace=True)
boone_covid_claims
```

```python
lake_covid_claims = pd.merge(Lake_weekly_claims[['Claims', 'Date']], 
                                           lake_covid[['Count', 'Date']], how='outer', on='Date', sort=True)
lake_covid_claims.fillna(0, inplace=True)
lake_covid_claims
```

```python
warrick_covid_claims = pd.merge(Warrick_weekly_claims[['Claims', 'Date']], 
                                           warrick_covid[['Count', 'Date']], how='outer', on='Date', sort=True)
warrick_covid_claims.fillna(0, inplace=True)
warrick_covid_claims
```

```python
boone_covid_claims.to_json("static/data/WeeklyBooneData.json")
lake_covid_claims.to_json("static/data/WeeklyLakeData.json")
warrick_covid_claims.to_json("static/data/WeeklyWarrickData.json")
```
